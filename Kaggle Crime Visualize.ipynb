```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from __future__ import division
import matplotlib as mpl
import zipfile
import os
#sns.set(style="whitegrid", color_codes=True)
%matplotlib inline
```

```python
#Load training data to look at

trainfile = "train"
datadir=r'C:\Users\Neil\Desktop\Kaggle Crime'
#file = os.path.join(datadir,trainfile)
```

```python
z = zipfile.ZipFile(os.path.join(datadir,trainfile)+".csv.zip")
df = pd.read_csv(z.open(trainfile+'.csv'), parse_dates=['Dates'])
```

```python
df['Year'] = df['Dates'].map(lambda x: x.year)
df['Month'] = df['Dates'].map(lambda x: x.month)
df['Hour'] = df['Dates'].map(lambda x: x.hour)

#create counter column
```

```python
#Lets look at the structure
df.head()
```

```python
#Category is our dependent variable and Dates
for i in df.ix[:,'Category':'Address']:
    df[i] = df[i].astype('category')

#relook at the structure much better
df.info()
```

```python
#Function that creates random gradient for barcharts
def random_gradient(data):
    
    dflen = len(data)
    end = np.random.randint(dflen+20,dflen+100)
    constant = np.random.uniform(.1,.9)
    
    #print dflen,end,constant
    return [(x/dflen,x/end,constant) for x in range(dflen)]
```

```python
category_count = df['Category'].value_counts().iloc[::-1]
title = 'test'
fontdict = {'fontsize': 18}
fig = plt.figure(figsize=(16,9))
fig.suptitle('Chicago Crime Count', fontsize=14, fontweight='bold')
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 12
#mpl.rcParams['title.labelsize'] = 16
category_count.plot(kind='barh', color=random_gradient(category_count), stacked=True)
```

```python
#plt.figure(figsize=(12,7))
mpl.rcParams['ytick.labelsize'] = 13
mpl.rcParams['xtick.labelsize'] = 12
monthly_events = df[['Month','Category']].groupby(['Month']).count().reset_index()
monthly_events.plot(kind='barh', stacked = True,colormap = "Blues", figsize=(12,6), legend = False, title ="Crime On Monthly Bases")
```

```python
events_perday = df[['Dates','Category']].query('Category == "LARCENY/THEFT"') .groupby(['Dates']).count()
events_perday = events_perday.resample('D',how=sum)
events_perday.dropna(inplace = True)
#events_perday = events_perday.set_index('Dates')

events_perday.tail()
```

```python
N = 2248
colors = np.linspace(0,1,N)
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 14
fig = plt.figure(figsize=(16,9))
fig.suptitle('Larceny/Theft by Day', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111,axisbg='black')
ax.grid(b=False)

#plt.xlim([2003-01-06,'2015-05-13'])
ax.scatter(events_perday.index, events_perday['Category'],c=colors ,cmap = "winter")
#plt.savefig('foo.png')
```
